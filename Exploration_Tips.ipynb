## Merging Data frames

```python
d1 = data[['Name', 'Location' , 'Fuel_Type']].sample(frac = 0.1, random_state = 1)
d2 = data[['Name', 'Owner_Type' , 'Mileage']].sample(frac = 0.1, random_state = 1)
```

```python
d1.merge(d2, how='inner', left_on='Name', right_on='Name')
```
